```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression


dataset = pd.read_csv(r'C:\Users\A Barisal\kc_house_data.csv')
space=dataset['sqft_living']
price=dataset['price']
x = np.array(space).reshape(-1,1)
y = np.array(price)
#print(x)
#dataset.head()
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size =.30, random_state = 0)

#from sklearn.linear_model import LinearRegression

Regressor = LinearRegression()
Regressor.fit(xtrain, ytrain)#creating the values
#transform putting those created values in the called function or appropriare oplace 

#x_test=linearRegression(m)
ped= Regressor.predict(xtest)
#y_answer=linearRegressor.predict(8.9)
#print(y_answer)
plt.scatter(xtrain, ytrain, color = 'red')
plt.plot(xtrain, Regressor.predict(xtrain), color = 'blue')
plt.title("Visuals for Test dataset")
plt.xlabel("space")
plt.ylabel("price")
plt.show()

plt.scatter(xtest, ytest, color = 'red')
plt(xtrain, Regressor.predict(xtrain), color = 'blue')
plt.title("Visuals for Test dataset")
plt.xlabel("space")
plt.ylabel("price")
plt.show()
```
